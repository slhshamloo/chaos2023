```python
import numpy as np
from matplotlib import pyplot as plt
plt.rc("text", usetex=True)
plt.rc('text.latex', preamble=r"\usepackage{amsmath}")
plt.rc("axes", labelsize=20) 
plt.rc("xtick", labelsize=12)
plt.rc("ytick", labelsize=12)

x2, x1 = np.mgrid[0:1:100j, 0:1:100j]
x1x2 = x1 * x2
xdot1 = x1x2 * (1 - 2 * x1)
xdot2 = x1x2 * (1 - 2 * x2)
speed = np.sqrt(xdot1**2 + xdot2**2)
lw = 5 * speed / speed.max()

plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.subplot().set_aspect("equal")
plt.streamplot(x1, x2, xdot1, xdot2, color="mediumblue", linewidth=lw,
               broken_streamlines=False)
plt.scatter([0.5], [0.5], c="red")
plt.savefig("6.4.10.f.pdf", bbox_inches="tight")
plt.show()
```
