```python
import numpy as np
from matplotlib import pyplot as plt
plt.rc("text", usetex=True)
plt.rc('text.latex', preamble=r"\usepackage{amsmath}")
plt.rc("axes", labelsize=20) 
plt.rc("xtick", labelsize=12)
plt.rc("ytick", labelsize=12)

x = np.linspace(0, 5, 10000)
r = 2*x / (1 + x**2)**2
s = x**2 * (1 - x**2) / (1 + x**2)**2

plt.xlabel(r"$r$")
plt.ylabel(r"$s$")
plt.subplot().set_aspect("equal")
plt.ylim(0, 0.15)
plt.plot(r, s, c="mediumblue")
plt.text(0.45, 0.025, "3 Fixed Points",
         ha="center", va="center", c="blueviolet")
plt.text(0.1, 0.025, "1 Fixed Point",
         ha="center", va="center", c="blueviolet")
plt.savefig("3.5.7.e.pdf", bbox_inches="tight")
plt.show()
```
