```python
import numpy as np
from scipy.integrate import solve_ivp
from matplotlib import pyplot as plt
plt.rc("text", usetex=True)
plt.rc('text.latex', preamble=r"\usepackage{amsmath}")
plt.rc("axes", titlesize=20, labelsize=20) 
plt.rc("xtick", labelsize=12)
plt.rc("ytick", labelsize=12)

s, r = np.mgrid[0:10:1000j, 0:10:1000j]

a_s, a_r, k = 1.2, 1, 15
rdot = -r + a_s + k * s * np.exp(-s)
sdot = -s + a_r + k * r * np.exp(-r)

speed = np.sqrt(rdot**2 + sdot**2)
lw = 3 * speed / speed.max()

def func(t, x, a_s, a_r, k):
    return -x[0] + a_s + k * x[1] * np.exp(-x[1]), -x[1] + a_r + k * x[0] * np.exp(-x[0])
sol = solve_ivp(func, (0, 15), (0, 0), args=(a_s, a_r, k),
                method="BDF", dense_output=True)
t = np.linspace(0, 15, 10000)
rsol, ssol = sol.sol(t)

plt.title(fr"$A_S={a_s},\quad A_R={a_r},\quad k={k}$")
plt.xlabel(r"$r$")
plt.ylabel(r"$s$")
plt.subplot().set_aspect("equal")
plt.streamplot(r, s, rdot, sdot, color="mediumblue", linewidth=lw,
               broken_streamlines=False)
plt.plot(rsol, ssol, color="red", lw=3)
plt.savefig("7.2.19.d.pdf", bbox_inches="tight")
plt.show()
```
